### Business Metrics

Since our goal is to get the best sales method for the new product line with less time spent on each customer, I would recommend we use the conversion rate as our metric for getting the best sales method.

Conversion Method = Number of porducts sold / Number of Visits by the customer on our website

Based on our 6 week data, the best method that made more sales with minimal effort was emailing first and then calling the customer on the third week which is the 'Email + Call' method. Therefore, if this method is increasing used for the majority of customers, it would significantly increase the number of sales.

### Recommendation

For the upcoming weeks, I would recommend we can focus on the following steps:

- Implementing 'Email + Call' & 'Email' sales methods to 60% and 30% of the total customers respectively. 
- Using the 'Email + Call' sales method should be our top priority for the majority of the customers since they have the highest revenue rates.
- Reducing the 'Call' sales method to only 5%-10% of the total customers or even less while ideally constituting mostly new customers in the 5%-10% of customers since the conversion rate for new customers is comparatively high and takes an average of 30mins for each call.
